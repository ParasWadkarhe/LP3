```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("C:/Users/Paras/Downloads/uber.csv")
df
```

```python
df = df.drop(['Unnamed: 0', 'key', 'pickup_datetime'], axis = 1)
df.dropna(inplace=True)
```

```python
sns.boxplot(df['fare_amount'])
```

```python
low = df['fare_amount'].quantile(0.05)
high = df['fare_amount'].quantile(0.95)
df = df[(df['fare_amount'] > low) & (df['fare_amount'] < high)]
```

```python
sns.boxplot(df['fare_amount'])
```

```python

plt.boxplot(df['fare_amount'])
```

```python
corr = df.corr()
sns.heatmap(corr, annot=True)
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(['fare_amount'], axis=1)
y = df['fare_amount']
X_train, x_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
RMSE = np.sqrt(mean_squared_error(y_test, y_pred))
print("RMSE -> ", RMSE)
R2 = r2_score(y_test, y_pred)
print("R2 -> ", R2)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train, y_train)
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
RMSE = np.sqrt(mean_squared_error(y_test, y_pred))
print("RMSE -> ", RMSE)
R2 = r2_score(y_test, y_pred)
print("R2 -> ", R2)
```

1. Importing libraries

pandas â†’ For handling and analyzing data in tables (dataframes).

seaborn â†’ For creating beautiful statistical plots.

matplotlib.pyplot â†’ For plotting graphs (basic plotting).

numpy â†’ For numerical operations like calculations, arrays, etc.

2. Loading the dataset
df = pd.read_csv('D:/SPPU BE Lab Practicals/ML/uber.csv')
df


Explanation:

pd.read_csv() â†’ Reads the CSV file containing Uber data into a table (dataframe).

df â†’ Stores the Uber dataset.

3. Dropping unnecessary columns and cleaning missing values
df = df.drop(['Unnamed: 0', 'key', 'pickup_datetime'], axis = 1)
df.dropna(inplace=True)


Explanation:

drop() â†’ Removes columns that are not needed for prediction.

Unnamed: 0 â†’ Usually an index column.

key â†’ Likely an ID or unique key (not useful).

dropna() â†’ Removes rows that have missing values, so the dataset is clean.

4. Plotting a boxplot to check outliers
sns.boxplot(df['fare_amount'])


Explanation:

Shows a boxplot for the fare_amount column.

Helps visualize outliers (very high or very low fares).

5. Removing extreme outliers
low = df['fare_amount'].quantile(0.05)
high = df['fare_amount'].quantile(0.95)
df = df[(df['fare_amount'] > low) & (df['fare_amount'] < high)]


Explanation:

quantile(0.05) â†’ 5% percentile â†’ small fare values.

quantile(0.95) â†’ 95% percentile â†’ very high fare values.

Keep only the rows where fare is between 5% and 95%, removing extreme values.

6. Visualizing cleaned fare data
sns.boxplot(df['fare_amount'])
plt.boxplot(df['fare_amount'])


Explanation:

Two ways to plot boxplots to check how fare_amount looks after removing outliers.

sns.boxplot â†’ prettier Seaborn version.

plt.boxplot â†’ basic Matplotlib version.

7. Checking correlation
corr = df.corr()

Explanation:

df.corr() â†’ Calculates how strongly each feature (column) is related to others.

sns.heatmap â†’ Shows a colored map of correlations.

annot=True â†’ Shows actual numbers on the heatmap.

Helps identify features most related to fare_amount.

from sklearn.model_selection import train_test_split

Explanation:

X â†’ All columns except fare_amount (these are the features we use to predict).

train_test_split â†’ Splits data into training and testing sets:

X_train, y_train â†’ Used to train the model.

x_test, y_test â†’ Used to check model performance later.

9. Linear Regression Model
from sklearn.linear_model import LinearRegression

y_pred = model.predict(x_test)


Explanation:

LinearRegression() â†’ A simple model that predicts fare as a straight line combination of features.

fit() â†’ Trains the model using training data.

predict() â†’ Uses the trained model to predict fares for test data.

10. Evaluating Linear Regression
from sklearn.metrics import mean_squared_error, r2_score

Explanation:

mean_squared_error â†’ Measures average squared difference between predicted and actual values.

np.sqrt() â†’ Convert MSE to RMSE (Root Mean Squared Error) â†’ same units as fare.

r2_score â†’ Measures how well the model explains the data (closer to 1 is better).
